#1. Praprocessing data


```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('bolauad.csv')
data
```

```python
data.columns = [ 'Email', 'Cidera', 'Nama', 'Nim', 'Prodi', 'NoHp', 'Posisi']
data
```

```python
def_Cidera = '-'
data.Cidera.fillna(def_Cidera, inplace = True)
data
```

```python
data.isna().sum()
```

#2. Encoding Bagian Posisi dan Cidera

```python
from sklearn.preprocessing import LabelEncoder
# # Encoding produk
# # Contoh data
# data = pd.DataFrame({'Posisi': ['Gelandang', 'Penyerang', 'Bek Tengah', 'Kiper', 'Bek Kanan','Striker']})

# Encoding
le = LabelEncoder()
data['Posisi'] = le.fit_transform(data['Posisi'])
data
```

```python
le = LabelEncoder()
data['Cidera'] = le.fit_transform(data['Cidera'])
data
```

```python
X = data[['Cidera', 'Posisi']]
X
```

#3. Menentukan Jumlah Cluster yang Sesuai dengan Elbow

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
# Menentukan jumlah cluster menggunakan metode elbow
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Visualisasi metode elbow
plt.plot(range(1, 11), wcss)
plt.title('Metode Elbow')
plt.xlabel('Jumlah Cluster')
plt.ylabel('WCSS')
plt.show()
```

#4. Hasil Cluster

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Menggunakan jumlah cluster optimal (misalnya 4 berdasarkan elbow)
optimal_clusters = 2
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
data['Cluster'] = kmeans.fit_predict(X)

# Visualisasi hasil clustering
plt.scatter(data['Posisi'], data['Cidera'], c=data['Cluster'], cmap='rainbow')
plt.title('Hasil Clustering Pemain Baru UKM SEPAKBOLA UAD')
plt.xlabel('Posisi')
plt.ylabel('Cidera')
plt.yticks(ticks=range(len(le.classes_)), labels=le.classes_)  # Menampilkan label produk
plt.show()

# Menampilkan hasil clustering
print(data[['Cidera', 'Posisi', 'Cluster']])
```

```python
from sklearn.metrics import silhouette_score

# Menghitung Silhouette Score
silhouette_avg = silhouette_score(X, data['Cluster'])
print(f'Silhouette Score: {silhouette_avg:.4f}')
```

#5.Menyimpan data bersih

```python
data.to_csv('data.csv', index=False)
```
